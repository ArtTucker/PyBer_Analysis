```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Set the x-axis to a list of strings
x_axis = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

# Set y-axis to a list of floats.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.grid()
plt.show()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Add the legend
plt.legend()
```

```python
# Create the plot, legend, etc.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Create the plot, legend, etc.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the y limit between 0 and 45.
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add error bars

# Add the legend
plt.legend()
```

```python
plt.bar(x_axis,y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the axes
plt.xlabel('Date')
plt.ylabel('Fare ($)')
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis,y_axis)
```

```python
# Create a properly aligned horizontal bar-chart
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create a properly aligned horizontal bar-chart
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
# Create labels for the axes
plt.xlabel('Fare ($)')
plt.ylabel('Date')
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Create the plot with ax.plt() - object oriented
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create a horizontal bar chart using object oriented interface
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.invert_yaxis()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Month')
```

```python
# Create a scatter plot
plt.plot(y_axis,x_axis, "o", c="red", label="Chicago")
plt.gca().invert_yaxis()
# Create labels for the axes
plt.xlabel('Fare ($)')
plt.ylabel('Date')
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis,y_axis,s=[i*3 for i in y_axis])
```

```python
# Create a scatter plot using object-oriented method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble plot using object oriented method
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis,s=[i*5 for i in y_axis],c="skyblue",alpha=0.8,edgecolors="black",linewidth=2,label="Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.invert_yaxis()
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Month')
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))

plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%')
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0.2, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

fig, ax = plt.subplots()
ax.pie(y_axis, 
       explode=explode_values, 
       colors=colors, 
       autopct='%1.1f%%', 
       shadow=True,
       startangle=90,
       counterclock=False,
       radius=4, 
       labels=x_axis)
plt.show()
```
